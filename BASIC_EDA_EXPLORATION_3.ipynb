```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("Data_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("Test_set.xlsx")
test_df.head()
```

```python
# combining two dataset as one

final_df = train_df.append(test_df)
final_df
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
# feature engineering process

final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[0])
final_df['Arrival_Minutes'] = final_df['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minutes'] = final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Dep_Time'] = final_df['Dep_Time'].apply(lambda x : x.split(':')[0])
```

```python
final_df['Dept_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_Minutes'] = final_df['Dep_Time'].str.split(':').str[1]
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.drop('Dep_Hour', axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df['Dept_Hour'] = final_df['Dept_Hour'].astype(int)
final_df['Dept_Minutes'] = final_df['Dept_Minutes'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop' : 0, '1 stop' : 1, '2 stops' : 2, '3 stops' : 3, '4 stops' : 4, 'nan' : 1})
```

```python
final_df['Duration_Hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
# label encoding for airline feature

from sklearn.preprocessing import LabelEncoder
sc = LabelEncoder()
```

```python
final_df['Airline'] = sc.fit_transform(final_df['Airline'])
final_df['Source'] = sc.fit_transform(final_df['Source'])
final_df['Destination'] = sc.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = sc.fit_transform(final_df['Additional_Info'])
```
