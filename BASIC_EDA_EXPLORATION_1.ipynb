```python
## another way of finding missing values in all columns of a dataset

[i for i in df.columns if df[i].isnull().sum() > 0]
```

```python
# using heatmap to find missing values or plot something

sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
# combining two dataframes in pandas using merge()

final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
final_df
```

```python
final_df.head(2)
```

```python
# to check datatypes

final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_index = final_df.Country.value_counts().index
country_index
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# creating pie chart - top 3 counties which uses zomato

plt.pie(country_values[:3], labels = country_index[:3], autopct = '%1.2f%%')
```

**OBSERVATIONS :**

**1. ZOMATO MAXIMUM TRANSACTIONS ARE FROM INDIA**

**2. AFTER THAN USA THEN FOLLOWED BY UK**

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating_count'})
ratings
```

**CONCLUSIONS**

**1. WHEN RATING IS BETWEEN 4.5 TO 4.9 --> EXCELLENT**

**2. WHEN RATING IS BETWEEN 4.0 TO 3.4 --> VERY GOOD**

**3. WHEN RATING IS BETWEEN 3.5 TO 3.9 --> GOOD**

**4. WHEN RATING IS BETWEEN 3.0 TO 3.4 --> AVERAGE**

**5. WHEN RATING IS BETWEEN 2.5 TO 2.9 --> AVERAGE**

**6. WHEN RATING IS BETWEEN 2.0 TO 2.4 --> POOR**

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = "Aggregate rating", y = "Rating_count", data = ratings)
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating_count", hue = 'Rating color', data = ratings, palette = ['Blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

**OBSERVATIONS**

**1. NOT RATED COUNT IS VERY HIGH**

**2. MAXIMUM NUMBER OF RATINGS ARE BETWEEN 2.5 TO 3.4**

```python
 # count plot

sns.countplot(x = "Rating color", data = ratings, palette = ['Blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green']) 
```
